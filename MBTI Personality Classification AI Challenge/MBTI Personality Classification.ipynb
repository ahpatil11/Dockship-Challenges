```python
train = train.fillna('the')
```

```python
test = test.fillna('the')
```

```python
newtest = test
```

```python
train = train.drop('idx', axis = 1)
newtest = test.drop('idx', axis = 1)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
train['type'] = le.fit_transform(train['type'])
```

```python
train.head()
```

```python
newtrain = train.iloc[:, 1:]
```

```python
newtrain.head(1)
```

```python
list1= [i for i in range(258)]
new_Index = [str(i) for i in list1]
newtrain.columns = new_Index
newtrain.head(5)
```

```python
' '.join(str(x) for x in newtrain.iloc[0,0:])
```

```python
posts = []
for row in range(0,len(newtrain.index)):
    posts.append(' '.join(str(x) for x in newtrain.iloc[row,0:]))
posts[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
countvector=CountVectorizer(ngram_range=(2,2))
traindataset=countvector.fit_transform(posts)
```

```python
randomclassifier=RandomForestClassifier()
randomclassifier.fit(traindataset,train['type'])
```

```python
test_transform= []
for row in range(0,len(newtest.index)):
    test_transform.append(' '.join(str(x) for x in newtest.iloc[row,0:]))
test_dataset = countvector.transform(test_transform)
predictions = randomclassifier.predict(test_dataset)
```

```python
solution = pd.DataFrame({'idx': test.idx, 'type': predictions})
solution.head()
```

```python
solution['type'] = le.inverse_transform(solution['type'])
solution.head()
```

```python
solution.to_csv('Solution1.csv', index = False)
```
